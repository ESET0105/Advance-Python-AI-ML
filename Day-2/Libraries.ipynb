```python
gaming_data = pd.read_csv('ign_scores.csv',index_col='Platform')
```

```python
gaming_data.head()
```

```python
plt.figure(figsize=(6,5))
sns.barplot(x=gaming_data['Action'],y=gaming_data.index)
plt.title('Avg rating of Action games on different platforms')
```

## heatmap

```python
plt.figure(figsize=(6,5))
sns.heatmap(gaming_data , annot=True)
plt.xlabel('genre')
plt.title('gamescore by genre')
```
