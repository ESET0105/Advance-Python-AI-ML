# Handelling Numerical features

## Missing Null values

```python
## 2 ways of handelling missing values
# 1> droppping  .drop methods
# 2> filling/Data Imputation

```

```python
car_data = pd.read_csv('car 1.csv')
```

```python
car_data.head()
```

```python
print('Missing value percentage')
(((car_data.isnull().sum())/len(car_data))*100).sort_values(ascending=False)
```

```python
car_data.info()
```

```python
data_clean = car_data.dropna()
```

```python
data_clean.isnull().sum()
```

## Data IMputation

```python
car_data['Engine Fuel Type'] = car_data.groupby('Model')['Engine Fuel Type'].transform(lambda x:x.fillna(x.mode().iloc[0] if not x.mode().empty else None))
```

```python
car_datag = car_data.groupby('Model')['Engine Fuel Type']
car_datag.head()
```

```python
car_data['Engine Fuel Type'].isnull().sum()
```

```python
car_data['Engine HP'] = car_data.groupby('Model')['Engine HP'].transform(lambda x:x.fillna(x.mean()))
```

```python
car_data['Number of Doors'] = car_data.groupby('Model')['Number of Doors'].transform(lambda x:x.fillna(x.mean()))
```

```python
car_data['Engine Cylinders'] = car_data.groupby('Model')['Engine Cylinders'].transform(lambda x:x.fillna(x.mean())) 
```

```python
car_data.isnull().sum()
```

## Rule Based IMputations

```python
car_data['Engine Cylinders']= car_data['Engine Cylinders'].fillna(car_data['Engine Cylinders'].mode())
```

```python
car_data['Engine HP']= car_data['Engine HP'].fillna(car_data['Engine HP'].mode()[0])
```

```python
car_data['Engine Cylinders'].mode()
```

```python
car_data['Engine Cylinders']= car_data['Engine Cylinders'].fillna(4.0)
```
