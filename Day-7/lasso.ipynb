```python
from sklearn.preprocessing import LabelEncoder
label = LabelEncoder()
def encoding_values(df):
    encoded_cols = []

    object_cols = df.select_dtypes(include=['object']).columns

    for cols in object_cols:
        df[cols] = label.fit_transform(df[cols])
        encoded_cols.append(cols)
    return encoded_cols

    

df2 = df.copy()
encoding_values(df2)
```

```python
def find_corr_encoded(df):
    high_corr_columns = []
    
    for col in df.columns:
          
        corr_value = df[col].corr(df['SalePrice'])
        
        # Check if absolute correlation is above threshold
        if abs(corr_value) > 0.3:
            high_corr_columns.append(col)
    
    return high_corr_columns
```

```python

def find_high_corr_with_price(df):
    
    numeric_cols = df.select_dtypes(include=['int64', 'float64']).columns
    high_corr_columns = []
    
    for col in numeric_cols:
          
        corr_value = df[col].corr(df['SalePrice'])
        
        # Check if absolute correlation is above threshold
        if abs(corr_value) > 0.5:
            high_corr_columns.append(col)
    
    return high_corr_columns

```

```python
def correlated(df,cols):
    new_columns = []
    for col in cols:
        if df['SalePrice'].corr(df[col]) >0.2:
            new_columns.append(col)

    return new_columns

new_list = correlated(df1,cols)
```

```python
X = df1.drop(['SalePrice'],axis=1)
Y = df1[['SalePrice']]
Y.head()
```

```python
from sklearn.model_selection import train_test_split
xtrain , xtest , ytrain , ytest = train_test_split(X,Y,test_size=0.25,random_state=40)
```

```python
from sklearn.linear_model import LinearRegression,Lasso,Ridge,ElasticNet
LR = LinearRegression()
LS = Lasso()
RD = Ridge()
EN = ElasticNet()
```

```python
def model(model,Xtrain,xtest,ytrain):
    md = model.fit(Xtrain,ytrain)
    ypred = md.predict(xtest)
    return ypred
```

```python
from sklearn.metrics import r2_score

LRpred = model(LR,xtrain,xtest,ytrain)
print('r2 score with Linear regression : ',r2_score(ytest,LRpred)*100,"%")

LSpred = model(LS,xtrain,xtest,ytrain)
print('r2 score with Lasso : ',r2_score(ytest,LSpred)*100,'%')

RDpred = model(RD,xtrain,xtest,ytrain)
print('r2 score with Ridge : ',r2_score(ytest,RDpred)*100,'%')

ENpred = model(EN,xtrain,xtest,ytrain)
print('r2 score with Elastic Net : ',r2_score(ytest,ENpred)*100,'%')

```
