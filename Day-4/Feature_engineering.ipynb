```python
import pandas as pd
```

## Feature Engineering

```python
house_data = pd.read_csv('california_housing_train.csv')
```

```python
house_data.describe()
```

## types of feature Engineering
### 1> Normalization
### 2> Binning
Bucketing/Saving in small bins


```python
house_data.head()
```

```python
# standerd scaling of MSRP column
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
house_data['population'] = scaler.fit_transform(house_data[['population']])
```

```python
house_data['population'].head()
```

```python
# min-max scaling on total_rooms column
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
house_data['total_rooms'] = scaler.fit_transform(house_data[['total_rooms']])
```

## Binning
### 1>equal width
### 2>Equal Frequency
### 3>custom

```python
## Binning (convert continous value into categorical bins) of equal width for median_income column
house_data['median_income_bin'] = pd.cut(house_data['median_income'], bins=4 , labels=['low','medium','high','very high'])
```

```python
print(house_data.median_income_bin.describe())
```

```python
#custom bins
bins = [15000,70000,150000,300000,600000]
#labels = ['low','medium','high','very high']
house_data['house_value_bin'] = pd.cut(house_data['median_house_value'], bins=bins , labels=['low','medium','high','very high'])
```
